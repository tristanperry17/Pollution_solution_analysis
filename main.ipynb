```python
#Imports & declarations
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import scipy.stats as st
from pathlib import Path
from sklearn import datasets, linear_model
```

```python
#Path CSV data, commented with source website
#https://datacommons.org/place/country/CIV
cotepop = 'Resources/worldbankCotepop.csv'
coteim = 'Resources/CoteCM.csv'

#https://datacommons.org/place/country/GHA
ghanapop = 'Resources/worldbankGhanapop.csv'
ghanaim = 'Resources/GhanaCM.csv'

#https://datacommons.org/place/country/SLE
sierrapop = 'Resources/worldbankSierrapop.csv'
sierraim = 'Resources/SierraCM.csv'

#https://datacommons.org/place/country/TGO
togopop = 'Resources/worldbankTogopop.csv'
togoim = 'Resources/TogoCM.csv'
```

```python
pm = [65,75,80]
rf = [8,4.3,2.1]
plt.plot(rf,pm)
from scipy.stats import linregress
linregress(rf, pm)
```

```python
#Read CSV files into DataFrames
cpop = pd.read_csv(cotepop)
cim = pd.read_csv(coteim)

gpop = pd.read_csv(ghanapop)
gim = pd.read_csv(ghanaim)

spop = pd.read_csv(sierrapop)
sim = pd.read_csv(sierraim)

tpop = pd.read_csv(togopop)
tim = pd.read_csv(togoim)

#Optional DataFrame check for column names
cpop
cim
```

```python
#Plotting population data together for comparison
plt.xlabel('Year')
plt.ylabel('Population')
plt.plot(cpop['label'], cpop['Total Population'], label = "Cote D'Ivoire")
plt.plot(gpop['label'], gpop['Total Population'], label = 'Ghana')
plt.plot(spop['label'], spop['Total Population'], label = 'Sierra Leone')
plt.plot(tpop['label'], tpop['Total Population'], label = 'Togo')
plt.legend(loc='upper right')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0)
plt.title('Population over Time')
```

```python
#Plotting Infant Mortality together for comparison
plt.xlabel('Year')
plt.ylabel('Infant Mortality (%)')
plt.plot(cim['label'], cim['Infant Mortality Rate'], label = "Cote D'Ivoire")
plt.plot(gim['label'], gim['Infant Mortality Rate'], label = 'Ghana')
plt.plot(sim['label'], sim['Infant Mortality Rate'], label = 'Sierra Leone')
plt.plot(tim['label'], tim['Infant Mortality Rate'], label = 'Togo')
plt.legend(loc='upper right')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0)
plt.title('Infant Mortality Rate per Year')
```

```python
# Costs of proposed solution per year for each country:
#In Sierra Leone, the program will cost $15 million per year.
#In Cote d'Ivoire, the program will cost $20 million per year.
#In Togo, the program will cost $6 million per year.
#In Ghana, the program will cost $20 million per year. 

#Calculate cost per person for given populations each year

spop['Cost per Capita'] = (15000000 / spop['Total Population'])
gpop['Cost per Capita'] = (20000000 / gpop['Total Population'])
tpop['Cost per Capita'] = (6000000 / tpop['Total Population'])
cpop['Cost per Capita'] = (20000000 / cpop['Total Population'])

#Plotting Cost of Implementation together 
plt.xlabel('Year')
plt.ylabel('Cost per Capita')
plt.plot(cpop['label'], cpop['Cost per Capita'], label = "Cote D'Ivoire")
plt.plot(gpop['label'], gpop['Cost per Capita'], label = 'Ghana')
plt.plot(spop['label'], spop['Cost per Capita'], label = 'Sierra Leone')
plt.plot(tpop['label'], tpop['Cost per Capita'], label = 'Togo')
plt.legend(loc='upper right')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0)
plt.title('Cost per Capita per Year')
```
